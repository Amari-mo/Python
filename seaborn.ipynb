```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Car data saved.xlsx')
df.head()
```

```python
Variables = ['Min Price (Lakh)','Max Price (Lakh)','Range (kmpl)','CC','Seats','Variants','Type', 'Ex-Showroom Price','RTO','Insurance','Other','Onroad Price']
```

```python
plt.figure(figsize=(15,12))#size of the squre
for i,var in enumerate(Variables,start=1):#placing them into column and rows
    plt.subplot(4,3,i)
    sns.scatterplot(x=df[var], y=df['Max Price (Lakh)'], alpha=0.7)
    plt.title(f'Scatter plot of{var} vs onroad Price')
    plt.xlabel(var)
    plt.ylabel('Target')          
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))#size of the squre
for i,var in enumerate(Variables,start=1):#placing them into column and rows
    plt.subplot(4,3,i)
    sns.histplot (df[var], bins=20,color='red',kde = True)
    plt.title(f'Histogram of{var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')          
plt.tight_layout()
plt.show()
```
